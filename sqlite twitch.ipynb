```python
#joins top 1000 streamers of 2021 with 2024 based on who is still in the top 1000
query = '''SELECT
s21.channel AS streamer,
s21.average_viewers AS avg_viewers_2021,
s24.avg_viewers_per_stream AS avg_viewers_2024,
s21.followers AS followers_2021,
s24.total_followers AS followers_2024
FROM streamers2021 s21
JOIN streamers2024 s24
ON LOWER(s21.channel) = LOWER(s24.name);'''
df = pd.read_sql_query(query, connection)
df
```

```python
#the number of streamers who kept their spot in top 1000 from 2021 to 2024
query = '''SELECT
CASE WHEN s21.channel IS NULL THEN 'New in 2024'
ELSE 'Returning from 2021'
END AS status,
COUNT(*) AS num_streamers
FROM streamers2024 s24
LEFT JOIN streamers2021 s21
ON LOWER(s24.name) = LOWER(s21.channel)
GROUP BY status;'''
df = pd.read_sql_query(query, connection)
df
```

Here we will clean the dataset and match the names that are in both datasets but recorded differently so we can try to get an accurate look at who is actually still in the top 1000

```python
query = '''SELECT
    LOWER(REPLACE(REPLACE(name, '_', ''), ' ', '')) AS clean_name
FROM streamers2024;
'''
df = pd.read_sql_query(query, connection)
df
```

```python
#Adding cleaned names to both databases
cursor.execute('''ALTER TABLE streamers2021
ADD COLUMN clean_name TEXT;''')

cursor.execute('''ALTER TABLE streamers2024
ADD COLUMN clean_name TEXT;''')

connection.commit()

cursor.execute('''UPDATE streamers2021
SET clean_name = LOWER(REPLACE(REPLACE(channel, '_', ''), ' ', ''));''')

cursor.execute('''UPDATE streamers2024
SET clean_name = LOWER(REPLACE(REPLACE(name, '_', ''), ' ', ''));''')

connection.commit()

```

Now we can go in and fuzzy match the names to see account for people who changed their usernames or appear differently in each dataset

```python
#these are names that are similar to each other in both datasets
#now we manually go through them and see which names are actually the same and then map them to each other so they appear the same in both datasets
from itables import show

query = '''SELECT s2021.channel as channel2021, s2024.name as name2024, s2021.clean_name as clean2021, s2024.clean_name as clean2024
FROM streamers2021 s2021
JOIN streamers2024 s2024
ON (
       s2024.clean_name LIKE '%' || s2021.clean_name || '%'
    OR s2021.clean_name LIKE '%' || s2024.clean_name || '%'
   )
WHERE clean2021 <> clean2024;'''
df = pd.read_sql_query(query, connection)
show(df)

```

Using this I can create a new table to standardize the names in both tables so they are represented in the same way

```python
cursor.execute('''drop TABLE if exists name_map;''') 

connection.commit()
```

```python
#create a table containing the mapped names
cursor.execute('''CREATE TABLE name_map (
               old_name TEXT PRIMARY KEY, 
               new_name TEXT);''') 

connection.commit()

cursor.execute('''INSERT INTO name_map (old_name, new_name) VALUES
               ('xqcow', 'xqc'),
               ('riotgames(riotgames)', 'riotgames'),
               ('gronkh', 'gronkhtv'),
               ('蝦愛橘子(shuteyeorange)', 'shuteyeorange'),
               ('老皮(mobilmobil)', 'mobilmobil'),
               ('餐餐自由配(blusewillyretry)', 'blusewillyretry'),
               ('skipnholive', 'skipnho'),
               ('接接(godjj)', 'godjj'),
               ('followgrubby', 'grubby'),
               ('嬌兔(zrush)', 'zrush'),
               ('亞洲統神(asiagodtonegg3be0)', 'asiagodtonegg3be0'),
               ('antoinedaniellive', 'antoinedaniel'),
               ('g4grevenant', 'revenant'),
               ('スタンミ(sutanmi)', 'sutanmi'),
               ('六希夫(qttsix)', 'qttsix'),
               ('jlbichouu', 'bichouu'),
               ('羅傑(roger9527)', 'roger9527'),
               ('超負荷(sam1268)', 'sam1268');''')

connection.commit()

#adds new column where the normalized names will go 
cursor.execute('''ALTER TABLE streamers2021 ADD COLUMN normalized_name TEXT;''')
cursor.execute('''ALTER TABLE streamers2024 ADD COLUMN normalized_name TEXT;''')

#adds the values to the column for both datasets
cursor.execute('''UPDATE streamers2021
               SET normalized_name = clean_name;''')

cursor.execute('''UPDATE streamers2024
               SET normalized_name = clean_name;''')

cursor.execute('''UPDATE streamers2021
               SET normalized_name = (
               SELECT new_name
               FROM name_map
               WHERE old_name = streamers2021.clean_name)
               WHERE clean_name IN (SELECT old_name FROM name_map);''')

cursor.execute('''UPDATE streamers2024
               SET normalized_name = (
               SELECT new_name
               FROM name_map
               WHERE old_name = streamers2024.clean_name)
               WHERE clean_name IN (SELECT old_name FROM name_map);''')

connection.commit()
```

Now the channel names that are supposed to be the same but appeared different in both tables should now be the same and we can get an accurate look at who is still in the top 1000. Counting the number of returning vs new top 1000 streamers we get

```python
query = '''SELECT
CASE WHEN s21.normalized_name IS NULL THEN 'New in 2024'
ELSE 'Returning from 2021'
END AS status,
COUNT(*) AS num_streamers
FROM streamers2024 s24
LEFT JOIN streamers2021 s21
ON s24.normalized_name = s21.normalized_name
GROUP BY status;'''
df = pd.read_sql_query(query, connection)
df
```

Below we can see all the people who kept their spot in the top 1000 alongside their cleaned names in each table

```python
query = '''SELECT DISTINCT
    s24.normalized_name,
    s24.clean_name AS name_2024,
    s21.clean_name AS name_2021
FROM streamers2024 s24
JOIN streamers2021 s21
ON TRIM(LOWER(s24.normalized_name)) = TRIM(LOWER(s21.normalized_name))
ORDER BY s24.normalized_name;
'''
df = pd.read_sql_query(query, connection)
show(df)
```
